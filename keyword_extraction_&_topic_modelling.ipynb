```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
data = pd.read_csv('/content/audiobooks.csv')
def clean_description(description):
    description=str(description)
    parts = description.split("\n\n", 1)
    return parts[1] if len(parts) > 1 else description
def preprocess_text(text):
    text = re.sub(r'\s+', ' ', str(text))
    text = re.sub(r'\W', ' ', text)
    text = text.lower()
    text = re.sub(r'\b(?:br|p|b)\b', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.strip()
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

data['cleaned_description'] = data['Description'].apply(clean_description)
data['cleaned_description'] = data['cleaned_description'].apply(preprocess_text)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load pre-trained transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(data['cleaned_description'].tolist(), convert_to_tensor=True)
```

```python
from nltk import pos_tag, word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

stemmer = PorterStemmer()

def extract_keywords(description, model, top_n=5):
    description_embedding = model.encode([description], convert_to_tensor=True)

    words = word_tokenize(description)

    meaningful_words = [word for word, pos in pos_tag(words)
                        if word.lower() not in ENGLISH_STOP_WORDS and pos.startswith(('NN', 'VB', 'JJ'))]

    if not meaningful_words:
        return []

    stemmed_words = [stemmer.stem(word) for word in meaningful_words]

    word_embeddings = model.encode(stemmed_words, convert_to_tensor=True)

    cosine_similarities = cosine_similarity(description_embedding.cpu(), word_embeddings.cpu()).flatten()

    top_indices = np.argsort(cosine_similarities)[::-1]
    unique_keywords = []
    seen_words = set()

    for index in top_indices:
        word = meaningful_words[index]
        stemmed_word = stemmed_words[index]
        if stemmed_word not in seen_words:
            unique_keywords.append(word)
            seen_words.add(stemmed_word)
        if len(unique_keywords) >= top_n:
            break

    return unique_keywords

data['keywords'] = data['cleaned_description'].apply(lambda x: extract_keywords(x, model, top_n=5))

for index, row in data.iterrows():
    print(f"Description {index + 1}:")
    print(f"Extracted Keywords: {', '.join(row['keywords'])}")
    print("\n")
```

```python
def display_topics(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic #{topic_idx + 1}:")
        print(" | ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
        print("\n")

feature_names = vectorizer.get_feature_names_out()

display_topics(lda, feature_names, n_top_words=5)
```

```python
from sentence_transformers import SentenceTransformer, util
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
import pandas as pd

nltk.download('punkt')

stemmer = PorterStemmer()
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
data['keywords_string'] = data['keywords'].apply(lambda x: ' '.join(x))
dtm = vectorizer.fit_transform(data['keywords_string'])
lda.fit(dtm)

def extract_keywords(text, model, top_n=5):
    description_embedding = model.encode([text], convert_to_tensor=True)
    words = word_tokenize(text)
    stemmed_words = list(set(stemmer.stem(word.lower()) for word in words))
    word_embeddings = model.encode(stemmed_words, convert_to_tensor=True)
    distances = util.pytorch_cos_sim(description_embedding, word_embeddings)
    keywords = sorted(zip(stemmed_words, distances[0]), key=lambda x: x[1], reverse=True)[:top_n]
    return [keyword[0] for keyword in keywords]

def predict_topic(keywords, vectorizer, lda_model):
    keywords_string = ' '.join(keywords)
    keywords_vector = vectorizer.transform([keywords_string])

    topic_distribution = lda_model.transform(keywords_vector)
    topic = topic_distribution.argmax()

    return topic
```

```python
def get_topic_words(topic_idx, lda_model, feature_names, n_top_words=5):
    """
    Get the top words for a given topic index from the LDA model.

    Args:
    - topic_idx: Index of the topic.
    - lda_model: Trained LDA model.
    - feature_names: List of feature names (words).
    - n_top_words: Number of top words to display.

    Returns:
    - List of top words for the specified topic.
    """
    topic_words = lda_model.components_[topic_idx]
    top_words_idx = topic_words.argsort()[-n_top_words:][::-1]
    return [feature_names[i] for i in top_words_idx]

topic_idx = 4
top_words = get_topic_words(topic_idx, lda, vectorizer.get_feature_names_out(), n_top_words=5)
```

```python
def audiobook_analysis_pipeline(description):
    keywords = extract_keywords(description, model, top_n=5)
    topic = predict_topic(keywords, vectorizer, lda)
    top_words = get_topic_words(topic, lda, vectorizer.get_feature_names_out(), n_top_words=5)

    return keywords, topic, top_words
```

```python
new_description = """
Oh darling, don’t be sorry that that man is dead. The one good thing he did was to be the inspiration for the charity work I do. I haven’t always done this. I used to be an actress, you know, a leading lady of the stage. Men would come from all over the city to see my shows, for I was beautiful then. Sometimes after the shows they would approach me, usually blind drunk, complimenting my dress or what was underneath it, asking for a kiss or a drink or a night. And sometimes they didn’t ask. My husband became my husband not because he asked but because he damaged me, and he was rich enough to get away with it and clever enough to know that marriage was my only option.
"""

keywords, topic, top_words = audiobook_analysis_pipeline(new_description)

print(f"Keywords: {keywords}")
print(f"Topic: Topic #{topic + 1}")
print(f"Top words for Topic #{topic + 1}: {', '.join(top_words)}")
```
